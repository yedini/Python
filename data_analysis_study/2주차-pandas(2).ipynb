## 1. pivot_table
 목적에 따라 행과 열의 위치를 바꾸거나 
 특정 요인에 따라 집계를 해서 구조를 바꿀 때 사용한다.

#### pd.pivot_table(data, index, columns, values, aggfunc)
 - index, columns: 행방향 또는 열방향에 표시할 열이름
 - values: 새 테이블의 값으로 표현할 열이름
 - aggfunc=np.sum / aggfunc=np.mean : 표시할 값이 중복될 경우 집계하는 함수

sales-funnel 파일을 불러와 name컬럼을 기준으로 pivot_table 명령 적용
(중복된 name들을 모두 하나로 표현한다)

```python
import numpy as np
import xlrd
df=pd.read_excel("sales-funnel.xlsx")
df.head()
```

```python
pd.pivot_table(df, index=["Name"])
```

```python
#다르게 재구성하기
pd.pivot_table(df, index="Name", columns="Account", values="Price")
```

 - 범죄데이터 (crime_anal_raw) 재구성하기

```python
crime_anal_raw=pd.read_csv("crime_in_Seoul_include_gu_name.csv", encoding='utf-8', index_col=0)
crime_anal_raw
```

```python
crime_anal=pd.pivot_table(crime_anal_raw, index='구별', aggfunc=np.sum)
crime_anal
```

```python
#각 범죄별 검거율 계산
crime_anal['강간검거율']=crime_anal['강간 검거']/crime_anal['강간 발생']*100
crime_anal['강도검거율']=crime_anal['강도 검거']/crime_anal['강도 발생']*100
crime_anal['살인검거율']=crime_anal['살인 검거']/crime_anal['살인 발생']*100
crime_anal['절도검거율']=crime_anal['절도 검거']/crime_anal['절도 발생']*100
crime_anal['폭력검거율']=crime_anal['폭력 검거']/crime_anal['폭력 발생']*100
del crime_anal['강간 검거']
del crime_anal['강도 검거']
del crime_anal['살인 검거']
del crime_anal['절도 검거']
del crime_anal['폭력 검거']
crime_anal.head()
```

#### np.sum()
 - 배열 원소 간 합치기 범용 함수.
 - axis=0 : 같은 열의 위+아래 원소 값을 더하여 1차원 배열 반환
 - axis=1 : 같은 행의 왼쪽+오른쪽 원소값을 더하여 1차원 배열 반환

 - 범죄 발생 열들의 합계 구하기

```python
col=['살인 발생', '강도 발생', '강간 발생', '절도 발생', '폭력 발생']
crime_anal_raw['범죄']=np.sum(crime_anal_raw[col], axis=1)
crime_anal_raw
```

 - crime_anal_raw를 구별 기준으로 피봇팅하고 중복되는 행은 합계 구하기

```python
crime_anal2=pd.pivot_table(crime_anal_raw, index="구별", aggfunc=np.sum)
crime_anal2
```

#### 실습: 인구 소멸 위기 지역 분석

- fillna() 함수 : 빈 셀에 대해 Nan처리를 하지 않고 그 앞 내용으로 채운다.

```python
population=pd.read_excel('population_raw_data.xlsx', header=1)
population.head()
```

```python
population.fillna(method='pad', inplace=True)
population.head()
```

```python
#column 이름 변경
population.rename(columns={'행정구역(동읍면)별(1)':'광역시도', '계':'인구수',
                  '행정구역(동읍면)별(2)':'시도'}, inplace=True)
population=population[(population['시도'] != '소계')]
population.head()
```
