# BEAUTIFULSOUP
- 웹크롤링을 할 때 사용되는 라이브러리.
- html 태그 등의 파일을 가져온 뒤 사용자가 parsing하기 쉽게 도와준다.
- html request => html response => html parsing

```python
from bs4 import BeautifulSoup
```

```python
#html 파일 불러오기
page=open('test_first.html', 'r').read()
soup=BeautifulSoup(page, 'html.parser')
print(soup.prettify())  #prettify(): 들여쓰기를 사용하여 보기좋게 출력함
```

#### .children : 한 단계 아래에 포함된 태그에 접속

```python
list(soup.children)
```

```python
html=list(soup.children)[2]
html
```

```python
#한단계 더 아래의 태그 접속
list(html.children)
```

```python
# 한단계 더 - body부분 접속
body=list(html.children)[3]
body
```

### html 태그 찾기

#### .find(태그, class=자료형, id=아이디,,,) 
- 제일 첫번째 해당 태그를 찾음

```python
soup.find('p')
```

#### .find_all(태그, class=자료형, id=아이디 ,,,)
 - 모든 해당 태그를 찾음

```python
soup.find_all(id='first')
```

```python
soup.find_all('p', class_='outer-text')
```

#### .next_sibling
 - 바로 다음 태그로 이동

```python
soup.head
```

```python
soup.head.next_sibling
```

```python
soup.head.next_sibling.next_sibling
```

```python
soup.body.p
```

```python
soup.body.p.next_sibling.next_sibling
```

#### .get_text(): 태그 안에 있는 텍스트만 출력

```python
for each_tag in soup.find_all('p'):
    print(each_tag.get_text())
```

#### 링크 찾기 : 'a' 태그

```python
soup.find_all('a')
```

```python
links=soup.find_all('a')
for each in links:
    href=each['href']
    text=each.string
    print(text + '->' + href)
```

### Requests / urllib
- html 문서 요청을 위한 library

```python
from urllib.request import urlopen
```

- urllib의 urlopen : url을 통해 해당 페이지를 읽어오는 함수

```python
##url변수에 주소를 저장하고 html을 읽어오기
url = "https://finance.naver.com/marketindex/"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
print(soup.prettify())
```

#### get('사이트 주소')

#### find_element_by_name('html_name')
- html의 name태그를 이용하여 특정 영역을 지정함

```python
#검색창 지정
search=driver.find_element_by_name('q')
```

#### send_keys()
- 지정된 영역에 원하는 문자를 입력함
#### submit()
- 엔터키와 같은 역할

```python
#검색창에 python 검색하기
search.send_keys('python')
search.submit()
```

#### find_element_by_xpath(‘/html/body/some/xpath')
- xpath를 이용하여 특정 영역을 지정함
- *xpath얻는 방법 : 해당 html 영역에서 마우스 오른쪽 버튼 -> copy xpath

```python
driver.get('http://www.opinet.co.kr/searRgSelect.do')
list=driver.find_element_by_xpath('//*[@id="SIGUNGU_NM0"]')
```

#### find_element_by_tag_names('h1')
- 특정 태그를 지정

```python
list_option=list.find_elements_by_tag_name('option')
```

#### get_attribute('속성')
- 특정 속성값을 얻음

```python
names=[option.get_attribute('value') for option in list_option]
names.remove('')
names
```

```python
#지역구 바꾸기
select=driver.find_element_by_id('SIGUNGU_NM0') #지역구 칸 지정
select.send_keys(names[0])
```
